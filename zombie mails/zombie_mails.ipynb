```python
x = open('Desura.txt',"r")
f = x.read()
arc = f.split('\n')
mails = [i  for i in arc if '@hotmail.com' in i]
```

```python
def mails(mails):
    mail = []
    for i in mails:
        try:
            m,p = i.split(":")
            mail.append(m)
        except:
            pass
    hotmails = []
    for i in mail:
        try:
            a,b = i.split("@")
            hotmails.append(a)
        except:
            pass
    return hotmail
```

```python
hotmails = mails(mails)
```
